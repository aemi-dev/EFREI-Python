## Exercice 1

 - 1) Create a function called `hypotenuse`, which takes two numbers as parameters and prints the square root of the sum of their squares.
 - 2) Call this function with two floats.
 - 3) Call this function with two integers.
 - 4) Call this function with an integer and a float.

```python
import math

def hypotenuse(x,y):
    print(math.sqrt(x**2+y**2))

hypotenuse(5.4,7.2)
hypotenuse(5,7)
hypotenuse(5,7.2)
```

## Exercice 2

 - 1) Rewrite the `hypotenuse` function from exercise 1 so that it returns a value instead of a display. 
 - 2) Call the function with two numbers and display the result.
 - 3) Call the function with two strings and display the result.
 - 4) Call the function with one number and one `string` and display the result.

```python
import math

def hypotenuse(x,y):
    if not(type(x) in (int,float) and type(y) in (int,float) ):
        return None
    return math.sqrt(x**2+y**2)

print(hypotenuse(5.4,7))
print(hypotenuse('5.4','7'))
print(hypotenuse(5.4,'7'))
```

## Exercice 3

 - 1) Write a function called `calculator`. It must take the following parameters: two numbers, an arithmetic operation (which can be addition, subtraction, multiplication or division and is addition by default), and an output format (which can be integer or floating point, and is floating point by default). The division must be a floating point division.
 
  The function must perform the requested operation on the two input numbers and return a result 

 - 2) Call the function with the following sets of parameters and check that the response is what you expect:
     - `2`, `3.0`
     - 2, `3.0`, the operation is division
     - 2, `3.0`, operation is division, output format is integer

```python
def calculator(x, y, operation='+', decimal=True):
    
    if operation == '+':
        z=x+y
    elif operation == '-':
        z=x-y
    elif operation == '*':
        z=x*y
    elif operation == '/':
        z=x/y
    else:
        return None

    if not decimal:
        z=int(z)

    return z
    


print(calculator(x=2, y=3.0))
print(calculator(x=2, y=3.0, operation='/'))
print(calculator(x=2, y=3.0, operation='/',decimal=False))

```

## Exercice 4


Rewrite the `calculator` function from the previous exercise so that it takes a *some number of numbers* and the same optional parameter( or parameters (you think about it)). The function should apply the operation to the first two numbers, then apply it again to the result and the next number, and so on. For example, if the numbers are 6, 4, 9 and 1 and the operation is subtraction, the function should return `6 - 4 - 9 - 1`. If only one number is entered, it should be returned as it is. 

```python
def calculator(liste_valeurs=[], operation='+', decimal=True):
   

    indice = 1
    resultat=liste_valeurs[0]

    while indice < len(liste_valeurs):
        if operation == '+':
            resultat += liste_valeurs[indice]
        elif operation == '-':
            resultat -= liste_valeurs[indice]
        elif operation == '*':
            resultat *= liste_valeurs[indice]
        elif operation == '/':
            resultat /= liste_valeurs[indice]
        else:
            return None
        indice += 1

    if not decimal :
        resultat=int(resultat)

    return resultat
    
  
print(calculator([1.5,2]))
print(calculator([1.5,2], decimal=False))
print(calculator([6,4,9,1], operation='-'))
print(calculator([2,3.0], operation='/'))
```
